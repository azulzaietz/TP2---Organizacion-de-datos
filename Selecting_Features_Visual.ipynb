```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
train = pd.read_csv('train_features.csv')
test = pd.read_csv('test_features.csv')
```

1) Realizamos una seleccion visual de los features. Pudimos observar que los features que serían buenos para la clasificación son: palabras_unicas, cant_stop_words, cant_url, prom_long_palabra, cant_puntuacion, cant_hashtags, cant_menciones, cant_signos_pregunta, cant_oraciones, cant_numeros, cant_mayusculas, cant_vocales ya que presentan mas diferencia en la distribución entre el Target verídico y no verídico por lo que serán más útiles para el aprendizaje del algoritmo de Machine Learning

```
FEATURES = ['cant_palabras', 'palabras_unicas', 'cant_stop_words', 'cant_url', 'prom_long_palabra', 'cant_puntuacion', 'cant_hashtag', 'cant_mencion', 'cant_signos_pregunta', 'cant_signos_exclamacion', 'subjetividad', 'cant_oraciones', 'cant_apariciones_keyword', 'cant_numeros', 'cant_minusculas', 'cant_mayusculas', 'cant_vocales', 'cant_consonant']

fig, axes = plt.subplots(ncols=2, nrows=len(FEATURES), figsize=(20, 50), dpi=100)
'cant_consonant'
for i, feature in enumerate(FEATURES):
    sns.distplot(train.loc[train['target'] == 0][feature], label='No verídico', ax=axes[i][0], color='green')
    sns.distplot(train.loc[train['target'] == 1][feature], label='Verídico', ax=axes[i][0], color='red')

    sns.distplot(train[feature], label='Train', ax=axes[i][1])
    sns.distplot(test[feature], label='Test', ax=axes[i][1])
    
    for j in range(2):
        axes[i][j].set_xlabel('')
        axes[i][j].tick_params(axis='x', labelsize=12)
        axes[i][j].tick_params(axis='y', labelsize=12)
        axes[i][j].legend()
    
    axes[i][0].set_title(f'{feature} Distribución del Target en el set de Entrenamiento', fontsize=13)
    axes[i][1].set_title(f'{feature} Distribución de los sets de Entrenamiento y Test', fontsize=13)

plt.show()
```
